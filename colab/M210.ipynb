Modelo de PPP:

## MAXIMIZE Z=20X1 + 12X2 +30X3 + 25X4

Sujeito A = 
#x1+x2+x3+x4 <=50 (mão de Obra) 
#3x1+2x2 +x3 <=100 (materia prima A)
#x2 + 2x3 + 3x4 <=90 (materia prima B)
#x1,x2,x3,x4 >=0


```python
from scipy.optimize import linprog
```

```python
#PARA MAXIMIZAÇÃO OS COEFICIENTES SERAO NEGATIVOS E PARA MINIMIZAÇÃO SERAO POSITIVOS 
objetivo = [- 20 , -12 , -40 , -25] # mazimização coloca - 

#COEFICIENTES DAS RESTRIÇÕES
coef_restricao = [[1, 1 , 1 , 1],
                  [3, 2 , 1 , 0],
                  [0, 1 , 2 , 3]]

lado_direito = [50,100, 90]   #limite (<= acima)

otimizador = linprog(c= objetivo, A_ub=coef_restricao , b_ub=lado_direito , method = 'revised simplex')
# C É o vetor com custos ou lucros individuais
# A_ub é a matriz com coeficintes das restrições
# b_ub é a matriz com os termos do lado direito
# method é o metodo usado para otimização (padrão = simplex)
```

```python
otimizador
# X = VALORES DE X 
```

###Exemplo 2
Modelo de PPP:

## MAXIMIZE Z=5X1 + 7X2

Sujeito A = 
#x1<=16 
#2X1 +3X2 <=19 
#X1+ X2 <=8 
#x1,x2 >=0

```python
import numpy as np
#from scipy.optimize import linprog
```

```python
objetivo2 = np.array([ -5 , -7])

coef_restricao2= np.array([[ 1, 0],
                           [2 , 3],
                           [1 , 1],
                           [-1, 0],  # esse e abaixo para nao negatividade 
                           [0 ,-1]])# quando usa numpy

lado_direito2 = np.array([16 , 19 , 8 , 0 , 0]) 
```

```python
resultado = linprog(objetivo2, coef_restricao2,lado_direito2)
```

```python
print(resultado.x)
```

###MINIMIZE z = 10x1 + 15x2 + 25x3

Sujeito A
*   x1 + x2 + x3 >= 1000(RESTRIÇÃO 1
*   x1 -2x2 >=0 (RESTRIÇÃO2
*   x3>= 340 (RESTRIÇÃO 3
*   X1,X2,X3 >=0  



#

```python
import numpy as np
#from scipy.optimize import linprog

objetivo3 = np.array([10 , 15 , 25])

coef_restricao3 = np.array([[-1 , -1 , -1],
                           [-1,  2 , 0],
                           [ 0 , 0 ,-1],
                           [-1 , 0 , 0],   # nao negatividade
                           [ 0 ,-1 , 0],  # nao negatividade
                           [ 0 , 0 ,-1]])  # nao negatividade

lado_direito3 = np.array([-1000 , 0 , -340, 0 , 0 , 0])
```

```python
resultado3 = linprog(objetivo3, coef_restricao3,lado_direito3)
```

```python
print("lucro otimo", resultado3.fun)
print("Ponto otimo", resultado3.x)
```
