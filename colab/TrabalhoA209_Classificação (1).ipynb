```python
import pandas as pd
from google.colab import files
import io
```

```python
uploaded = files.upload()
dados = pd.read_csv (io.StringIO(uploaded['survey lung cancer.csv'].decode('utf-8')))
```

```python
dados
```

```python
# Separação dos dados de entrada e saida 
saida = dados['LUNG_CANCER']
entrada = dados.drop(columns ='LUNG_CANCER')
entrada


```

```python
#separação dos dados para treino e teste
from sklearn.model_selection import train_test_split
entradas_treino,entradas_teste,saidas_treino,saida_teste = train_test_split(entrada,saida)

```

```python
# DEFINIÇÃO E TREINAMENTO DE REDE NEURAL
from sklearn.neural_network import MLPClassifier
rede_neural = MLPClassifier(hidden_layer_sizes=(5,4),activation='logistic',max_iter=900)
rede_neural.fit (entradas_treino,saidas_treino)
```

```python
# Estimar os valores da saida da RNA quando colocadas as entradas de teste 
saidas_estimadas = rede_neural.predict(entradas_teste)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(saida_teste,saidas_estimadas))
print(confusion_matrix(saida_teste,saidas_estimadas))
```
